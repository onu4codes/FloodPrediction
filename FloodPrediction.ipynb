```python
import pandas as pd
import numpy as np
flood=pd.read_csv('/content/drive/MyDrive/train.csv')
flood
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
flood_scaled=scaler.fit_transform(flood)
flood=pd.DataFrame(flood_scaled,columns=flood.columns)
flood.shape
```

```python

x=flood.iloc[:,1:-1]
y=flood.iloc[:,-1]
display(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,y_pred)
r2
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=5)
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
r2=r2_score(Y_test,y_pred)
r2
```

```python

from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=20,max_features=10,min_samples_split=20,splitter='best')
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
r2=r2_score(Y_test,y_pred)
r2
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(max_depth=8,max_features=15,min_samples_split=20,n_estimators=100)
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
r2=r2_score(Y_test,y_pred)
r2
```
